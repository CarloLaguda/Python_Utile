```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1) Importare, pulire dai valori nulli e visualizzare le prime 5 righe del dataset (1pt)

```python
df = pd.read_csv("high_popularity_spotify_data.csv")
```

```python
df.isnull().sum()
```

```python
df_pulito = df.dropna()
df_pulito.to_csv("high_popularity_spotify_data_clean.csv", index = False, sep = ",")
```

```python
df_pulito.head()
```

### 2) Crea un grafico a barre che mostri il numero di brani per ciascun playlist_genre (1pt)

```python
plt.figure(figsize=(9, 5))
cxg = df_pulito.groupby("playlist_genre").size()
cxg.plot(kind = "bar")
plt.title("Bumero di brani per genere")
plt.xlabel("Genere")
plt.ylabel("Canzoni")
plt.show()
```

### 3) Crea un grafico a barre della media di valence per ciascun playlist_genre. (1pt)

```python
plt.figure(figsize=(9, 5))
mxg = df_pulito.groupby("playlist_genre")["valence"].mean()
mxg.plot(kind = "bar")
plt.title("Valence media per genere")
plt.xlabel("Genere")
plt.ylabel("Valence")
plt.show()
```

### 4) Top 5 artisti con più brani (1pt)

```python
plt.figure(figsize=(9, 5))
tar = df_pulito.groupby("track_artist").size().sort_values(ascending = False).head()
tar.plot(kind = "barh")
plt.title("Valence media per genere")
plt.xlabel("Genere")
plt.ylabel("Valence")
plt.show()
```

### 5) Crea un grafico a linee che confronti i valori medi di speechiness e acousticness per ciascun playlist_genre (2pt)

```python
plt.figure(figsize=(9, 5))
plt.plot(df_pulito.groupby("playlist_genre")["speechiness"].mean(), label = "speechiness" )
plt.plot(df_pulito.groupby("playlist_genre")["acousticness"].mean(), label = "acousticness" )
plt.title("Valori medi speechiness e acousticness")
plt.ylabel("Valore medio")
plt.xlabel("Playlist")
plt.legend()
plt.show()
```

### 6) Crea un grafico a linee che mostri il numero di brani per ciascun playlist_subgenre (1pt)

```python
plt.figure(figsize=(9, 5))
cxg = df_pulito.groupby("playlist_subgenre").size()
cxg.plot(kind = "line")
plt.title("Bumero di brani per genere")
plt.xlabel("Genere")
plt.ylabel("Canzoni")
plt.show()
```

### 7) Crea un grafico a torta che mostri i 5 sottogeneri (playlist_subgenre) più rappresentati nel dataset. (1pt)

```python
plt.figure(figsize=(9, 5))
spr = df_pulito.groupby("playlist_subgenre").size().sort_values(ascending = False).head()
plt.pie(spr, autopct='%1.1f%%', startangle = -180)
plt.show()
```
