#### 1. Trasformare il file cvs allegato su classroom in un dataframe

#### 2. Visualizzare le prime 7 righe del dataframe

```python
dataframe.head(7)
```

#### 3. Visualizzare i prodotti con origine Olandese

```python
dataframe[dataframe["origin"] == "Olanda"]
```

#### 4. Visualizzare solo i prodotti con prezzo di vendita superiore a 5

```python
dataframe[dataframe["salePrice"] > 5]
```

#### 5. Visualizzare i prodtti con origine Olandese con prezzo di vendita superiore a 5

```python
dataframe[(dataframe["salePrice"] > 5) & (dataframe["origin"] == "Olanda")]
```

#### 6. Quanti prodotti ci sono per ogni paese?

```python
dataframe.groupby("origin").size()
```

#### 7. Qual'è il prezzo medio di vendita dei prodotti per ogni paese?

```python
dataframe.groupby("salePrice").mean()
```
