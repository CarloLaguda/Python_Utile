```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
```

```python
df = pd.read_csv("dataset.csv")
df
```

## 1) Mostra la spesa media per ogni città, ordinando le città dalla spesa più bassa alla più alta.

```python
plt.figure(figsize=(9, 5))
spp = df.groupby("Città")["Spesa_mensile"].mean().sort_values(ascending = True)
spp.plot(kind = "bar")
plt.title("Spesa media per paese")
plt.xlabel("Città")
plt.ylabel("Media")
plt.show()
```

## 2) Visualizza lo stesso grafico dell'esercizio 1 usando plt.barh al posto di plt.bar

```python
plt.figure(figsize=(9, 5))
spp = df.groupby("Città")["Spesa_mensile"].mean().sort_values(ascending = True)
spp.plot(kind = "barh")
plt.title("Spesa media per paese")
plt.xlabel("Città")
plt.ylabel("Media")
plt.show()
```

## 3) Visualizza la spesa media per ogni categoria di acquisti

```python
plt.figure(figsize=(9, 5))
spc = df.groupby("Categoria_acquisti")["Spesa_mensile"].mean()
spp.plot(kind = "bar")
plt.title("Spesa media per categoria")
plt.xlabel("Categoria")
plt.ylabel("Media")
plt.show()
```

## 4) Visualizza il numero di clienti per ciascuna città 

```python
plt.figure(figsize=(9, 5))
abp = df.groupby("Città").size()
abp.plot(kind = "bar")
plt.title("Spesa media per categoria")
plt.xlabel("Categoria")
plt.ylabel("Media")
plt.show()
```

## 5) Scatter plot - Mostra la relazione tra età e spesa mensile.

```python
eta = df["Età"]
sm = df["Spesa_mensile"]
plt.figure(figsize=(9, 5))
plt.scatter(eta, sm, color= "red")
plt.title("Relazione tra eta e spesa media")
plt.xlabel("Eta")
plt.ylabel("Spesa Media")
plt.show()
```

## 6) Trova il modo di visualizzare la spesa totale per città suddivisa per categoria come fatto in figura

```python
x = df.groupby(["Città", "Categoria_acquisti"])["Spesa_mensile"].sum().unstack()
x.plot(kind = "bar", stacked = True)
plt.figure(figsize=(9, 5))
```

## 7) Rappresenta la percentuale di spesa totale suddivisa per categoria di acquisti.

```python
dati = df.groupby("Categoria_acquisti")["Spesa_mensile"].sum()
plt.pie(dati, autopct='%1.1f%%', startangle = -180)
plt.show()
```

## 8) Visualizza la percentuale di clienti soddisfatti (4 e 5 stelle) e insoddisfatti (1, 2 e 3 stelle)

```python
s = df[df["Soddisfazione"]>=4]
i = df[df["Soddisfazione"]<=3]

x = len(s)
y = len(i)
sizes = [x, y]

plt.pie(sizes, autopct='%1.1f%%', startangle = -180 )
plt.show()
```
