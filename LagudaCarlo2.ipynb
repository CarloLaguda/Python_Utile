```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### 1) Importare, pulire dai valori nulli e visualizzare le prime 5 righe del dataset (0.5pt)

```python
df = pd.read_csv("amazon_sales_data 2025.csv")
df.isnull().sum()
```

```python
df_pulito = df.dropna()
df_pulito.to_csv("amazon_sales_data 2025_pulito.csv", index = False, sep = ",")
```

```python
df_pulito.head()
```

### 2) Creare un grafico a linee che mostri le vendite totali (Total Sales) per ogni prodotto (Product) (0.5pt)

```python
x = df_pulito.groupby("Product").sum("Total Sales")
vtp = x["Total Sales"]
vtp
vtp.plot(kind = "line")
plt.show()
```

#### 3) Visualizzare le vendite totali (Total Sales) per ogni categoria di prodotto (Category) usando un grafico a barre. (0.5pt)

```python
x = df_pulito.groupby("Category").sum("Total Sales")
tsc = x["Total Sales"]
tsc.plot(kind = "bar")
plt.show()
```

#### 4) Creare un grafico a barre che mostri il totale delle vendite (Total Sales) per ogni Customer Location. (0.5pt)

```python
x = df_pulito.groupby("Customer Location").sum("Total Sales")
tscl = x["Total Sales"]
tscl.plot(kind = "bar")
plt.show()
```

#### 5) Creare un grafico a barre che mostri i 5 clienti che hanno speso di più (1pt)

```python
x = df_pulito.groupby("Customer Name").sum("Total Sales").sort_values(ascending = False).head()
tsc = x["Total Sales"]
tsc.plot(kind = "bar")
plt.show()
```

#### 6) Creare un grafico a torta che mostri la distribuzione percentuale dello stato degli ordini (Status) (1pt)

```python
x = df_pulito.groupby("Status").size()
plt.pie(x, autopct='%1.1f%%', startangle = -180)
plt.show()
```

#### 4) Creare un grafico a doppia linea che confronti per ogni categoria di prodotto (Category) il prezzo medio (Price) e la quantità totale venduta (Quantity) (2pt)

```python
x = df_pulito.groupby("Category")["Price"].mean()
y = df_pulito.groupby("Category")["Quantity"].sum("Quantity")


plt.plot(df_pulito.groupby("Category")["Price"].mean(), label='Prezzo Medio', marker='o')
plt.plot(df_pulito.groupby("Category")["Quantity"].sum("Quantity"), label='Quantità vendute', marker='s')
plt.legend()
plt.show()
```

#### 8) Creare un grafico a barre raggruppate che mostri, per ciascun prodotto, il numero di ordini in base allo stato (Status), confrontando ad esempio quanti ordini sono "Completed", "Pending", "Cancelled" per ciascun Product (2pt)

#### 9) Che cos’è uno scatter plot? In quali casi è utile utilizzarlo? Descrivi come si crea uno scatter plot in Python usando matplotlib.pyplot, indicando la funzione da utilizzare e un esempio pratico. (1pt)

```python
Lo scatter plot (o grafico a dispersione) è un tipo di grafico che ci permette di poter rappresentare la relazione tra due 
variabili quantitative (numeriche). Diventa utile usarlo quando dobbiamo analizzare la relazione che c'è tra due tipi di dati.'
Per creare uno scatter plot in python dobbiamo prima importare la libreria matpoltlib.pyplot, poi posssiamo creare il notro
scatter plot usando il metodo: plt.scatter(variabile1, variabile2).
Un esempio pratico di come possiamo usare uno scatter plot è per analizzare le ore di studio in relazione ai voto finale: 
```

```python
voto = [10, 7, 9, 8, 6]
ore_studio = [4, 1, 2, 3, 5]
plt.scatter(voto, ore_studio)

plt.show()
```
