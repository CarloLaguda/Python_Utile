### Per esempio....

```python
# aggiungiamo una nuova colonna in base alle vendite
# diciamo che se la colonna Global_Sales Ã¨:
# maggiore uguale di 1 allora ha venduto tanto
# altrimenti ha venduto poco
df_pulito.loc[df['Global_Sales'] >= 1, 'vendite'] = 'Tanto'
df_pulito.loc[df['Global_Sales'] < 1, 'vendite'] = 'Poco'
df_pulito
```
